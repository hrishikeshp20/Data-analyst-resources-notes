```python
x = np.linspace(0,5,11)
y = (x ** 2)
```

```python
#fig = plt.figure(figsize=(length,width), dpi=100) -- creates a blank canvas
#axes = fig.add_axes([left, bottom, width, height]) #enter numbers

#STEP1
fig = plt.figure()  #Creates blank canvas  

#STEP2
#Set dimensions graphs inside blank canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #value range should be between 0&1
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

#Plot values
axes1.plot(x,y,'-r') 
axes1.set_title('Larger Plot')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2.plot(y,x,'b')
axes2.set_title('Smaller Plot')
axes2.set_ylabel('Y')
axes2.set_xlabel('X')

#zooms in to the above plot and gets only the plot from values mentioned below
axes2.set_xlim([0,1]) 
axes2.set_ylim([0,2])
```

```python
#works same as above but below line is combination of STEP1 & STEP2
#fig,axes = plt.subplots(nrows=4,ncols=4,figsize=(length,width), dpi=100) #4 rows and 4 columns
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout() #fixes overlapping 

#Use indexing to plot
# GRAPH-1
axes[0,0].plot(x,y, color='orange') # colour hex code also works in color
axes[0,0].set_title('First plot')

#GRAPH-2
axes[0,1].plot(y,x, linewidth = 10, alpha = 0.3, marker='o',markersize=10) 
axes[0,1].set_title('second plot')                           #alpha for transparency, range(0-1)
axes[0,1].set_ylabel('Y')                                    #linewidth or lw
axes[0,1].set_xlabel('X')                                    #marker - shows where each points lies on the line


#GRAPH-3
axes[1,0].plot(x,x**2, label = 'X Squared') #label is required for axes.legend() function
axes[1,0].plot(x,x**3, label = 'Y Cubed')
axes[1,0].set_title('second plot')
axes[1,0].set_ylabel('Y')
axes[1,0].set_xlabel('X')
axes[1,0].legend(loc = 0) #legend-location
                          #loc is location for legend in graph. read documentation for 
                          #http://matplotlib.org/users/legend_guide.html
        
#GRAPH-4
axes[1,1].plot(y,x, linestyle=':') #alpha for transparency, range(0-1)
axes[1,1].set_title('second plot')               #linestyle or ls
axes[1,1].set_ylabel('Y')
axes[1,1].set_xlabel('X')
#ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```
