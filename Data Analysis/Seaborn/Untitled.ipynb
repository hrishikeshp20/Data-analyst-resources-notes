```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
# seaborn has its own dataset
```

```python
tips
```

```python
Displot -- univariable(one variable)
#additional arguements
# kde =True/False
#bins = 12 Number of bars
sns.displot(tips['total_bill'], bins = 20)
```

```python
#JOINT PLOT -- Combination of 2 displot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#https://seaborn.pydata.org/generated/seaborn.jointplot.html
```

```python
sns.pairplot(tips, hue='sex')
#creates jointplot for each and every column
#In hue pass categorical value column
#https://seaborn.pydata.org/generated/seaborn.pairplot.html
```

```python
sns.rugplot(tips['total_bill'])
#same as displot but instead of bars it plots lines
```

```python
sns.kdeplot(tips['total_bill'])
#plots kde
```

```python
CATEGORICAL PLOTS
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.sum)
#basically visualisation of groupby
#groupby on x so pass category column in x
#can pass any function in estimator, default is average
```

```python
sns.countplot(x='sex', data=tips)
#sane as barplot but just shows count
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#x = categorical column
#y = numeric
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
#x = categorical column
#y = numeric
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue ='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
#combination of stripplot and violin plot
```

```python
#can stack plots on each other
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='bar')
#General plot
#Use kind to get desired graph
```
